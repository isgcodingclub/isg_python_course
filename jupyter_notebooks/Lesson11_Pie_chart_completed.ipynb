<h1> Lesson 11: Pie charts</h1>
1. Simple pie chart. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

<h3> Pie chart parameters</h3>
<ul> <li> data itself - a list. The only compulsory parameter</li><li> colors, a list of colors. There is also an option to use pre-made seaborn palettes.  </li><li> labels of the data - the list with labels</li><li> hatch - list with hatches</li><li> radius - radius of the pie chart. Deafult is 1.</li><li>startangle, the angle at which the first wedge is rotated. Deafult is 0.</li><li>autopct : The percentage label<ul><li>autopct = '%.1f' - display the percentage value to 1 decimal place</li><li>autopct = '%.2f' display the percentage value to 2 decimal places</li><li>autopct = '%.1f%%' display the pecentage to 1 decimal place and display percentage sign</li><li>autopct = '%.2f%%' display the pecentage to 2 decimal places and display percentage sign</li></li></ul></li><li>wedgeprops - it is a dictionary that contains the properties of the wedges<ul><li>linewidth - number, width of the line</li><li>linestyle : "-", "-.", "--"</li><li>'antialiased' True or False. If True wedge line is smooth</li><li>Edgecolor - color of the edges</li></li></ul></li><li>explode - the list that defines which slice is going to "explode"</li></li></ul>



```python
data = [30, 40, 30]
fig, ax = plt.subplots()
ax.pie(data, colors = ["#e88b84", "#857fdb", "#b3f0a8"], autopct = '%.0f%%', labels = ["label 1", "label 2", "label 3"], 
       hatch = ["//", "-.", "o"], radius = 0.6, startangle = 60, 
       wedgeprops = {'linewidth': 3, 'linestyle': '-', 'antialiased': True, "edgecolor":"b"}, explode = [0.2, 0.1, 0])
plt.legend(ncols = 3, loc = "lower center")
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("output_our_precious.csv")
df.head(10)
```

Exercise solution: 
<ol> <li> Dropping the NAs from the CDR2 column</li></ol>

```python
# drop the Na values
df_mod = df.dropna(subset = ["CDR2 length"])
# convert the column to list
list_all_values = df_mod["CDR2 length"].to_list()
# generate the list of the unique values using loop !!!
unique_value_list = [] # list to store uniques
for number in list_all_values:
    if number not in unique_value_list:
        unique_value_list.append(number)
print(unique_value_list)
# calcultaing the counts
list_of_counts = []
for value in unique_value_list:
    frequency = list_all_values.count(value)
    list_of_counts.append(frequency)
print(list_of_counts)
```

```python
# define the colors 
colors_list = sns.color_palette("pastel")
fig, ax = plt.subplots()
ax.pie(list_of_counts, labels = unique_value_list, colors = colors_list, wedgeprops = {'linewidth': 1, 'linestyle': '-', 'antialiased': True, "edgecolor":"k"}, autopct = '%.0f%%'  )
plt.legend(bbox_to_anchor = (-0.2, 1.2))
plt.show()
```
