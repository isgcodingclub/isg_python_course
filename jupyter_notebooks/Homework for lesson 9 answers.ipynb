<p><b>Exercise 2:</b> Plot the overall survival months against the patient age at diagnosis as a <b>scatter plot</b></p>

```python
import pandas as pd
df = pd.read_csv("METABRIC_RNA_Mutation.csv")
df2 = df[["patient_id", "age_at_diagnosis", "overall_survival_months"]]
df2.head(10)
```

```python
import matplotlib.pyplot as plt
x_data = df2["age_at_diagnosis"].to_list()
y_data = df2["overall_survival_months"].to_list()
fig, ax = plt.subplots()
ax.set_title("The scatter graph")
ax.scatter(x = x_data, y = y_data, marker = "D", color = 'r', edgecolors = 'k', s = 20)
plt.show()
```

<p><b> Exercise 3:</b> You are given two lists one of which contains the time and the second the OD. Please plot the line graph with title, axis labels and legend. Be creative. </p>

```python
time = [0, 2, 4, 6, 8, 10, 12, 14]
OD = [0.005, 0.075, 0.284, 0.315, 0.472, 0.530, 0.736, 0.901]
# make your graph below
fig, ax = plt.subplots()
ax.set_title("OD versus time", fontfamily = "serif", loc = "center")
ax.set_ylabel("OD", fontfamily = "serif")
ax.set_xlabel("time (days)", fontfamily = "serif")
ax.grid(linestyle = ":", linewidth = 0.5, color = "k")
ax.plot(time, OD, color = 'r', linewidth = 0.7)
ax.legend(labels = ["Bacterial growth"], edgecolor = "r")
plt.show()
```
