<h1> Lesson 7 : Modules </h1>

Modules in Python are simply Python files with the <code>.py</code> extension, which implement a set of functions. Modules are imported from other modules using the <code class = "python "> import</code> command.
<p> Let's consider the simple script that inverts the string </p>
<p> The scripts contains one function called inversion. I will show how to import this script into current script and call the function </p>

````python
def invert(string):
    return string[::-1]
def length_mes(string):
    return len(string)
````

````python
import inversion # importing the script
res = inversion.invert("Any string")
print(res)
````

````python
res2 = inversion.length_mes("AGT")
print(res2)
````

Sometimes the script name itself is too long. So it is imported using the alias name.

````python
import inversion as inv # importing the script
res = inv.invert("My_precious_string") # calling the function
print(res)
````

You can also import everything from the scripts by using an asterix.

````python
from inversion import * # imported all the functions from the script
print(invert("The_one"))
````

Slicing the dataframe using the <code> .iloc </code> method.
The syntax is
```python
dataframe.iloc[something]
```
where the "something can be :

- An integer, e.g. 5.

- A list or array of integers, e.g. [4, 3, 0].

- A slice object with ints, e.g. 1:7.

Unlike .loc method, which is <b> label </b> based the <code> .iloc </code> method is <b> number </b> based.

If you put in the square bracket <b> just one integer </b> that leads to the pandas series object creation

````python
import pandas as pd
df = pd.DataFrame({"Antibody": ["NRL-N-E2", "NRL-N-C2", "NRL-N-C2-hop"],
                   "CD": [62,67.5, 65], "Affinity": [0.8, 1, "Unknown"]})
df
````

````python
slice1 = df.iloc[1]
slice1
````

````python
type(slice1)
````

````python
slice2 = df.iloc[[1]]
slice2
````

Iterating through the dataframe using the iloc method.

````python
import pandas as pd
df = pd.DataFrame({"Antibody": ["NRL-N-E2", "NRL-N-C2", "NRL-N-C2-hop"],
                   "CD": [62,67.5, 65], "Affinity": [0.8, 1, "Unknown"]})
df
````

Iterating through one column at time using <code> .iloc </code> method.

````python
for i in range(len(df)):
  affinity = df.iloc[i, 2]
  print(affinity)
````

You can also iterate like that over the sub-section pof the dataframe, but the resultant object in each iteration will be the <b> datafrme </b>

````python
for i in range(len(df)):
  row = df.iloc[[i], [1, 2]]
  print(row)
  print(type(row))
````

Iterating over one or multiple columns by converting each column to the list.
1. First you need to convert your column to the list.
2. Iterate through the list

````python
list_affinity = df['Affinity'].to_list()
list_affinity1 = df.iloc[:, 2].to_list() # semicolon means everything, all rows from column 2.
print(list_affinity)
print(list_affinity1)
for a in list_affinity:
  if a == 0.8:
    print("Affinity is " + str(a))
  elif a != 0.8:
    print("Affinity is not 0.8, it is " + str(a))
````

Likewise you can iterate over multiple lists by converting multiple columns to list and then using the <code> zip() </code> method

````python
list_affinity3 = df['Affinity'].to_list()
list_antibody = df['Antibody'].to_list()
for aff, ant in zip(list_affinity3, list_antibody):
  print(ant + " : " + str(aff))
````

Filtered by multiple columns. There are several ways how you can do it. First is to use <code> df.query</code>. Filtering by partial string. Filter out the values based on the certain substring. The syntaxis is shown below. This method works well with the letters, but it might not work for special character.
```python
df[df["colum name"].str.contains("string")]
```

````python
import pandas as pd
df = pd.DataFrame({"Antibody": ["NRL-N-E2", "NRL-N-C2", "NRL-N-C2-hop"],
                   "CD": [62,67.5, 65], "Affinity": [0.8, 1, "Unknown"], "Source species": ["Llama","Camel", "Alpaca"]})
````

````python
df_filter = df[df["Antibody"].str.contains("C2")]
df_filter
````
