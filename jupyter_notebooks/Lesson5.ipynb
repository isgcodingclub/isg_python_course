- functions
> function is a piece of code that is executed only when it is callled.  You can pass data, known as parameters, into a function. A function can return data as a result.

function syntax:

```python
def function_name(parameters):
  do something with the parameters to generate output
  return output
```
To be executed the function needs to be called:
```python
function_name(parameters)
```

````python
def get_the_square(number):
  answer = number**2
  return answer
````

````python
# calling the function
print(get_the_square(3))
print(get_the_square(10))
````

### How to define the parameters for the function? ###

<br> There might be numerous parameters within the function, you can set them as <b> positional arguments </b> </br>
In tis example the length is first parameter, the width is second. So wen you put the number arguments in function wnhen yiu call it the values get assigned to the variable based on their positions.

````python
# Let's define the functions that calculates the area of the square
def area(length, width):
  area_value = length*width
  return area_value

area_calculated = area(40, 50)
print(area_calculated)
print(area(40, 50))
````

````python
def raise_to_power(to_be_raised, power):
  answer = to_be_raised**power
  return answer
print(raise_to_power(2, 3))
print(raise_to_power(3, 2))
````

Another way how to define arguments is to use **keyword arguments**. In this case the order does not matter.

````python
def raise_to_power(to_be_raised, power):
  answer = to_be_raised**power
  return answer

print(raise_to_power(power = 3, to_be_raised=2))
````

### Arbitary arguments *args ###
There are cases when the function <b> will not know </b> how many arguments it will receive. In this case you can put a <b>*</b> before your arguments and the function will receive a tuple of arguments, that you can handle within function.

````python
def ratio_first_to_rest(first, *rest):
  sumfind = sum(rest) # calculating the sum of the all arguments except first
  ratio = first/sumfind # calculation what is the ratio of the first to rest
  return ratio

print(ratio_first_to_rest(1, 10, 90))
print(ratio_first_to_rest(1, 10, 90, 20))
````

### Keyword arbitary arguments **Kwargs ###

Function does not know how many keyworld arguments it will receive, but you can define the function in a way so that it handles only some of them. For instance, in the example below the function only processes CDR1.

````python
def CDR_measure(**CDR):
  len_CDR1 = len(CDR["CDR1"])
  return len_CDR1

measured_CDR1_len = CDR_measure(CDR1 = "AASARTFYTMG", CDR2 = "GAIRWGVYATT", CDR3 = "AARAGPLGFELSATSSAEYDY")
print(measured_CDR1_len)
````

Default parameter value. Sometimes you may wish to specify one of the parameters when you define the function. For instance in the example below the function has default argument **treshold**

````python
def CDR_measure1(CDR, treshold = 5):
  output = ""
  len_CDR = len(CDR)
  if len_CDR > treshold:
    output += "long enough"
  else:
    output += "short"
  return output
print(CDR_measure1("GAIRWGVYATT"))
print(CDR_measure1("GAIR"))
````

You can return multiple things in your as a result of thje function. Below is the example of such function.

````python
def CDR_len_and_cysteine(CDR):
  cysteine_out = ""
  CDR_len = len(CDR)
  if CDR.find("C") == -1:
    cysteine_out = cysteine_out + "no cysteine"
  else:
    cysteine_out = cysteine_out + "there is cysteine"
  return CDR_len, cysteine_out

CDR_calculated, cysteine_status = CDR_len_and_cysteine("AARAGPLGFELCSATSSAEYDY")

print(CDR_calculated)
print(cysteine_status)
````
