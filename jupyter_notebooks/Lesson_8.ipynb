```python
import pandas as pd
```

```python
df = pd.read_csv("METABRIC_RNA_Mutation.csv")
df.head(10)
```

As we can see the dataset is too large. We do not need all that information. First exercise would be how to generate the smaller dataframe using only some selected columns the column names are available in the list below. <b> Please generate the smaller dataframe using the column names provided. </b>
and name it df1 please. 

```python
needed_columns = ["patient_id", "age_at_diagnosis", "chemotherapy", "overall_survival", "mutation_count", "overall_survival_months"]
df1 = df[needed_columns]
```

df1

<p><b>Dropping the empty rows.</b></p>
In the <code>mutation count</code> column some values are missing, there is a way how to drop them. 
1. The code below shows how to <b>drop the rows in which at least one element is missing</b>: 

```python
df2 = df1.dropna()
```

```python
df2
```

There is also an option to drop the columns in which <b> at least one element is missing </b>. This will obviously drop <code> mutation_count </code> column. 

```python
df3 = df1.dropna(axis='columns')
```

```python
df3
```

The code below shows how to <b> drop the rows in which all values are missing.</b> 
In our case that not going to derop anything, because we do not have such a rows. 

```python
df4 = df1.dropna(how='all')
df4
```

Keep only the rows with at least specified number of non-NA values is present. 

```python
df5 = df1.dropna(thresh=2) # keep the rows that have at least 2 non -NA values. 
df5
```

You can also specify the columns in which you want to look for NA values. 

```python
df6 = df1.dropna(subset = ["mutation_count", "overall_survival_months"])
```

```python
df6
```

If you have empty cells in your dataframe, you may wish not to drop the corresponding rows, but just replace the NAN values with the zeros. You can do it using the following command. This will replace all NANS with zeros. 

```python
df7 = df1.fillna(0)
df7
```

<b> Sorting the dataframe according to the certain criteria.</b> <p> Below is the way how to sort the dataframe accoeding the certain column. </p>

```python
df_sorted1 = df1.sort_values(by=['age_at_diagnosis'])
df_sorted1
```

```python
df3_sorted = df1.sort_values(by=['age_at_diagnosis'], ascending = False)
df3_sorted
```

```python
df_sorted2 = df1.sort_values(by=['age_at_diagnosis', 'overall_survival_months'])
df_sorted2
```
